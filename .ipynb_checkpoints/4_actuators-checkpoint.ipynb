# Actuators and Joint Controllers
The goal of this notebook is to implement a simulation of a simple electrical motor actuator and feedback controllers.

The dynamic of the actuator will be enriched to better reflect experiment performed with the Open Dynamic Robot Initiative test motor test bed.

## Set up
We will simply need Numpy and Matplotlib

```python

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
%config Completer.use_jedi = False
```

## A Motor Model 
To implement a motor model, we will create a class with a function to udpate the motor state given a control input. Internal variable of this class will correspond to the model state to be defined (position, velocity etc..) as well as model parameters.

```python
class Motor:
    def __init__(self,dt):
        self.dt = dt
        #State variables
        self.q = 0.0 #position de l'axe
        self.dq = 0.0 #vitesse de l'axe
        self.ddq = 0.0 #acceleration de l'axe
        self.tau = 0.0 #couple
        
        #Model Parameters
        self.Kv = 1.0
        self.J = 1.0
        self.retard = 10
        
        self.previous_i = [0.0 for i in range(self.retard)]

    def reset(self):
        '''Reset all the internal state variable'''
        self.q = 0.0
        self.dq = 0.0
        self.ddq = 0.0
        self.tau = 0.0
        
        
    def update(self,i):
        #todo
        self.previous_i.append(i)
        self.tau = self.Kv * self.previous_i[0]
        self.previous_i.pop(0)
        self.ddq = 1/self.J*self.tau
        self.dq += self.ddq*self.dt
        self.q += self.dq*self.dt                      
```

## Define a feedback controller
Together, we will design and test several feedback controller to servo our actuator in position. 

```python
mot = Motor(dt=0.001)
```

```python
mot.reset()
q = []
dq = []
ddq = []
l_i = []
Kp = 10
Kd = 8
#q_d = np.sin(np.linspace(0.,7,10000))
#dq_d = np.cos(np.linspace(0.,7, 100000))
q_d = 1.0
iq = 0
for j in range(10000):
    e = q_d - mot.q
    i = e*Kp - Kd*mot.dq + Ki*(q_d*j*mot.dt - iq)
    #i = min(0.5, i)
    #i = max(-0.5, i)
    mot.update(i)
    l_i.append(i)
    iq += mot.q*mot.dt
    q.append(mot.q)
    dq.append(mot.dq)
    ddq.append(mot.ddq)

    
print(q_d-mot.q)
```

```python
plt.plot(q_d)
plt.plot(q)
plt.show()
```

To have a function depending on interactive sliders

```python
@interact(frequency=(0,5,0.5)  ,  amplitude=(0,1.0,0.1))
def draw_sinwave(frequency,amplitude):
    t = np.linspace(0.,1.,100)
    data = amplitude*np.sin(2*np.pi*frequency*t)
    plt.plot(t,data)
    plt.show()
```
